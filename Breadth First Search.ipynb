# Breadth First Search

```python
import collections                                                 #using collections package to implement queue
def bfs(graph, root):                                              #start of function bfs with graph and root as params
    queue = collections.deque([root])                              #initializing queue with element roor
    visited = set()                                                #creating a set called visited
    while queue:                                                   #runs a loop until queue is empty
        vertex = queue.popleft()                                   #popping the latest element in queue into vertex
        visited.add(vertex)                                        #adding the vertex to the visisted array
        for i in graph[vertex]:                                    #loop to traverse through the neighbours
            if i not in visited:                                   #checks if i is present in visited array
                queue.append(i)                                    #adding the value if not present in visited
    print(visited)                                                 #printing the visisted array, our solution

if __name__ == "__main__":
    #Dictionary
    graph = {0:[1, 2, 3], 1:[0, 2], 2:[0, 1, 4], 3:[0], 4:[2]}     #our graph as a dictionary
    bfs(graph,0)                                                   #calling the bfs function
```
