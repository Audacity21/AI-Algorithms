# Depth First Search

DFS using Recursion

```python
graph = {'A': ['B','C','D'], 'B': ['E'], 'C': ['D','E'], 'D': [], 'E': []}       #our graph to be traversed
visited = set()                                                                  #initialising with empty set

def dfs(visited, graph, root):                                                   #function definition
    if root not in visited:                                                
        print(root)
        visited.add(root)                                                        #adding root to set if not visited
        for neignbour in graph[root]:                                            #traversing through neignbours
            dfs(visited, graph, neignbour)                                       #calling the function recursively

dfs(visited, graph, 'A')
```

```python
graph = {'A': ['B','C','D'], 'B': ['E'], 'C': ['D','E'], 'D': [], 'E': []}

def dfs_non_recursive(graph, source):
    if source is None or source not in graph:
        return "Invalid input"
    path = []
    stack = [source]
    while(len(stack) != 0):
        s = stack.pop()
        if s not in path:
            path.append(s)
        if s not in graph:
            #leaf node
            continue
        for neighbor in graph[s]:
            stack.append(neighbor)
    return " ".join(path)

DFS_path = dfs_non_recursive(graph, "A")
print(DFS_path)
```

# DFS in Trees

```python
graph = {"A":["B","C","D"], "B":["E"], "C":["F","G"], "D":["H"], "E":["I"], "F":["J"]}

def dfs_non_recursive(graph, source):
    if source is None or source not in graph:
         return "Invalid input"
    path = []
    stack = [source]
    while(len(stack) != 0):
        s = stack.pop()
        if s not in path:
            path.append(s)
        if s not in graph:
            #leaf node
            continue
        for neighbor in graph[s]:
            stack.append(neighbor)
    return " ".join(path)

DFS_path = dfs_non_recursive(graph, "A")
print(DFS_path)
```
